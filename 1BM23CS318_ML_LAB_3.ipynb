```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


df = pd.read_csv('/content/housing_area_price.csv')


X = df[['area']]
y = df['price']


model = LinearRegression()
model.fit(X, y)


print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)


predicted = model.predict(X)


plt.scatter(X, y)
plt.plot(X, predicted, color='red')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression - Using Library")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('/content/housing_area_price.csv')

X = df['area'].values
y = df['price'].values


X_matrix = np.c_[np.ones(len(X)), X]


theta = np.linalg.inv(X_matrix.T.dot(X_matrix)).dot(X_matrix.T).dot(y)

intercept = theta[0]
slope = theta[1]

print("Intercept:", intercept)
print("Slope:", slope)


y_pred = intercept + slope * X

plt.scatter(X, y)
plt.plot(X, y_pred, color='green')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression - Without Library")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/housing_area_price.csv')


X = df[['area']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

print("Coefficient (Area):".replace(':', '', 1), model.coef_[0])
print("Intercept:", model.intercept_)

prediction = model.predict([[3000]])
print("Predicted Price for area 3000:", prediction[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/housing_area_price.csv')

X = df[['area']].values
y = df['price'].values

X_matrix = np.c_[np.ones(X.shape[0]), X]

theta = np.linalg.inv(X_matrix.T.dot(X_matrix)).dot(X_matrix.T).dot(y)

print("Parameters (theta):")
print("Intercept:", theta[0])
print("Area Coefficient:", theta[1])

new_house = np.array([1, 3000])
predicted_price = new_house.dot(theta)

print("Predicted Price for area 3000:", predicted_price)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/hiring.csv')

df['experience'] = df['experience'].fillna(0)

word_to_num = {
    'zero':0, 'one':1, 'two':2, 'three':3, 'four':4,
    'five':5, 'six':6, 'seven':7, 'eight':8,
    'nine':9, 'ten':10, 'eleven':11, 'twelve':12
}

df['experience'] = df['experience'].replace(word_to_num).astype(int)

# Fill missing test scores
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mean()
)

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

predictions = model.predict([[2,9,6], [12,10,10]])

print("Salary for candidate 1:", predictions[0])
print("Salary for candidate 2:", predictions[1])
```
